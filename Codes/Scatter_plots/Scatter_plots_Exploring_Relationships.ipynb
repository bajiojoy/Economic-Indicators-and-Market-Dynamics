```python
import pandas as pd
import plotly.express as px

#Loading relevent data sets
sp500 =pd.read_csv('C:/Users/SHOZ/Desktop/Projects/Python/Economy_Perfomance/Data_sets/processed_data/clean_monthly_opening_500_data.csv')
m2 =pd.read_csv('C:/Users/SHOZ/Desktop/Projects/Python/Economy_Perfomance/Data_sets/processed_data/clean_M2_data.csv')
cpi = pd.read_csv('C:/Users/SHOZ/Desktop/Projects/Python/Economy_Perfomance/Data_sets/processed_data/clean_inflation_data.csv')
fed_funds = pd.read_csv('C:/Users/SHOZ/Desktop/Projects/Python/Economy_Perfomance/Data_sets/processed_data/clean_fed_funds_rate.csv')
gdp = pd.read_csv('C:/Users/SHOZ/Desktop/Projects/Python/Economy_Perfomance/Data_sets/processed_data/clean_GDP_data.csv')
un_employment = pd.read_csv('C:/Users/SHOZ/Desktop/Projects/Python/Economy_Perfomance/Data_sets/processed_data/clean_unemployment_data.csv')

```

```python
# Scatter plot for 'S&P 500 Volume' vs 'Money Supply' with trend line and markers
fig1 = px.scatter(sp500, x='Volume', y=m2['M2SL'], log_x=True, log_y=True,
                  title='S&P 500 Volume vs Money Supply',
                  labels={'Volume': 'S&P 500 Volume', 'y': 'Money Supply(M2)'},
                  trendline_color_override='black',  # Set trend line color to black
                  trendline='ols')  # Use ordinary least squares (OLS) for the trend line

# Update marker properties
fig1.update_traces(marker=dict(color='orange', size=5))  # Set marker color and size

fig1.update_layout(
    plot_bgcolor='rgba(0, 0, 0, 0)')  # Set the background color of the plot

fig1.write_html('interactive_Scatter_plot_S&P500_Volume_vs_Money_Supply.html')
fig1.show()
```
